```python
# Importing Libraries

import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize'] = 8, 16
```

##### Downloading the detection model now!

```python
reader = easyocr.Reader(['en'])
```

Importing Image

```python
Image("/content/drive/MyDrive/Datasets/Reading text from img/gangstawall.png")
```

```python
text_extracted = reader.readtext("/content/drive/MyDrive/Datasets/Reading text from img/gangstawall.png")
```

## Text generated is shown in 3 parts below :



1.    Co-ordinates of the Boundary Box.
2.   'The Text Extracted'
3.     Accuracy of Prediction.



```python
text_extracted
```

```python
coordinate = text_extracted[1][0]
```

```python
x_min, y_min = [int(min(idx)) for idx in zip(*coordinate)]
```

```python
x_max, y_max = [int(max(idx)) for idx in zip(*coordinate)]
```

### Now, using Matplotlib, let's see where the text 'CALM' is located.

```python
image = cv2.imread("/content/drive/MyDrive/Datasets/Reading text from img/gangstawall.png")
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
